```python
# Delete any duplicate entry in the dataset
alv_news_df.drop_duplicates(inplace=True)
alv_news_df.reset_index(drop=True, inplace=True)

# Delete any non-English news
from langdetect import detect
alv_news_df = alv_news_df[alv_news_df['title'].apply(lambda x: detect(x) == 'en')]
alv_news_df.reset_index(drop=True, inplace=True)

alv_news_df.info()
```

```python
# Explode along ticker_sentiment
alv_news_df['ticker_sentiment'] = alv_news_df['ticker_sentiment'].apply(lambda x: eval(x)) # should return a dict
alv_news_df = alv_news_df.explode('ticker_sentiment', ignore_index=True)
alv_news_df.reset_index(drop=True, inplace=True)
alv_news_df.head()
```

```python
alv_news_df.shape
```

```python
# Apply necessary transformations
# Extract the ticker and the sentiment label for each row
alv_news_df['ticker'] = alv_news_df['ticker_sentiment'].apply(lambda x: x['ticker'])
alv_news_df['sentiment'] = alv_news_df['ticker_sentiment'].apply(lambda x: x['ticker_sentiment_label'])

# Keep only the tickers we are interested in
alv_news_df = alv_news_df[alv_news_df['ticker'].isin(tickers)]
alv_news_df.reset_index(drop=True, inplace=True)

# Drop any duplicate ticker-headline pair
alv_news_df.drop_duplicates(subset=['ticker', 'title'], inplace=True)
alv_news_df.reset_index(drop=True, inplace=True)

# Annotate the title
alv_news_df["annot_title"] = "[TICKER] " + alv_news_df["ticker"] + " | [TITLE] " + alv_news_df["title"]

# The dataset use "Bearish", "Somewhat-Bearish", "Neutral", "Somewhat_Bullish", "Bullish" as sentiment labels
# We want to map the middle 3 to neutral, and rename the other 2 to positive and negative
alv_news_df['sentiment'] = alv_news_df['sentiment'].map(
    {'Bearish': 'negative',
     'Somewhat-Bearish': 'neutral',
     'Neutral': 'neutral',
     'Somewhat_Bullish': 'positive',
     'Bullish': 'positive'}
)

alv_news_df.head()

```

```python
# Create a function to map time_published to the right date
# If the news is released before 8pm UTC (i.e. 4pm EST), it counts as today. Else, it counts as the next day
# This is because we'll use the closing price for our prediction, and major stock exchanges close at 4pm EST
def time_to_date(time_published):
    time_published = pd.to_datetime(time_published)
    if time_published.hour <= 20:
        return time_published.date()
    else:
        return time_published.date() + pd.Timedelta(days=1)
```

```python
# Apply the function
alv_news_df['date'] = alv_news_df['time_published'].apply(time_to_date)
alv_news_df[['date', 'time_published']].head(30)
```

```python
# Check the range of dates included in the dataset
print(alv_news_df['date'].min(), alv_news_df['date'].max())
print(alv_news_df['date'].max() - alv_news_df['date'].min()+pd.Timedelta(days=1)) # inclusive
```

```python
# Check the number of dates with at least 1 headline in the dataset
alv_news_df['date'].value_counts()
```

The sentiment feature is likely a new feature for Alpha Vantage API as well, given that only news from 2024-12-11 were retrieved, even when the limit of 1000 was not reached for each ticker.

It is also noteworthy that every date in the dataset has at least 1 piece of news! However, given that there are some dates with less than 5 news pieces, meaning at least 1 ticker is not covered, a daily analysis might not be the most fitting

```python
# Explode the dataset by the "tickers" column
news_2024_df["tickers"] = news_2024_df["tickers"].apply(lambda x: eval(x)) # Should transform str to dict
news_2024_df = news_2024_df.explode('tickers', ignore_index=True)
news_2024_df.head()
```

```python
# Create a function to extract the sentiment label
def extract_sentiment(sent_dicts_str, ticker):
    if pd.isna(sent_dicts_str):
        return None
    sent_dicts = eval(sent_dicts_str)
    for sent_dict in sent_dicts:
        if sent_dict['ticker'] == ticker:
            return sent_dict['sentiment']
    return None
```

```python
# Apply necessary transformations
# Rename "tickers" to "ticker"
news_2024_df.rename(columns={"tickers": "ticker"}, inplace=True)

# Keep only the tickers we are interested in
news_2024_df = news_2024_df[news_2024_df['ticker'].isin(tickers)]
news_2024_df.reset_index(drop=True, inplace=True)

# Drop any duplicate ticker-headline pair
news_2024_df.drop_duplicates(subset=['ticker', 'title'], inplace=True)
news_2024_df.reset_index(drop=True, inplace=True)

# Extract the sentiment label for each row
news_2024_df['sentiment'] = news_2024_df.apply(lambda x: extract_sentiment(x['insights'], x['ticker']), axis=1)

# Annotate the title
news_2024_df["annot_title"] = "[TICKER] " + news_2024_df["ticker"] + " | [TITLE] " + news_2024_df["title"]

news_2024_df.head()
```

```python
# Count the number of entries in each label
news_2024_df['sentiment'].value_counts()
```

```python
# Keep only positive, neutral and negative entries
news_2024_df = news_2024_df[news_2024_df['sentiment'].isin(['positive', 'neutral', 'negative'])]
news_2024_df.reset_index(drop=True, inplace=True)
news_2024_df.head()
```

# 2. Train-test split

To keep consistent data density across time, I will combine `alv_news_df` (date range: 2024-12-11 to 2025-03-29) with headlines from 2024-12-07 (a Saturday, which affects next week's stock price) of `news_2024_df` as the test set for the sentiment analysis task, as well as the input data for the stock prediction model. The rest of `news_2024_df` will be merged with `news_2023_df` to train the sentiment classification model

```python
# Create a new column in each dataset indicating the name of the dataset it is sourced from
alv_news_df['source_dataset'] = 'alv_news_raw'
news_2024_df['source_dataset'] = 'polygon_news_2024_raw'
news_2023_df['source_dataset'] = 'polygon_news_2023_raw'
```

```python
alv_news_df.head()
```

## 2.1. Merge test set

```python
# Concat alv_news_df and news_2024_df along: ticker, sentiment, title, annot_title, date, source_dataset
merge_columns = ["ticker", "sentiment", "title", "annot_title", "date", "source_dataset"]

merged_val_df = pd.concat([alv_news_df[merge_columns], news_2024_df[news_2024_df['date'] >= pd.to_datetime("2024-12-07").date()][merge_columns]], axis=0)
merged_val_df.head()
```

```python
merged_val_df.shape[0]
```

```python
# If a pair of annot_title in alv_news_raw and polygon_news_2024_raw are identical, I keep the one from polygon_news_2024_raw (sourced from Polygon API) for maximum source consistency
merged_val_df = merged_val_df.sort_values(by="source_dataset", ascending=False) # This makes sure the duplicate headline from polygon_news_2024_raw always comes first
merged_val_df.drop_duplicates(subset=['annot_title'], inplace=True)
merged_val_df.dropna(subset=['sentiment'], ignore_index=True, inplace=True)
merged_val_df = merged_val_df.sort_values(by="date", ascending=True) # Sort by date in ascending order
merged_val_df.reset_index(drop=True, inplace=True)
merged_val_df.head()
```

```python
merged_val_df.shape[0]
```

```python
# Set out 1 example for each class from the validation set to use as one-shot classification
one_shot_df = merged_val_df.groupby('sentiment').head(1)
one_shot_df
```

```python
# Remove the examples from the validation set
merged_val_df = merged_val_df[~merged_val_df.index.isin(one_shot_df.index)]
merged_val_df.reset_index(drop=True, inplace=True)
merged_val_df.head()
```

```python
# Save the datasets
one_shot_df.reset_index(drop=True, inplace=True)
one_shot_df.to_csv('data/merged/news_one_shot.csv', index=False, encoding="utf-8")
merged_val_df.to_csv('data/merged/news_SA_test.csv', index=False, encoding="utf-8")
one_shot_df.shape[0], merged_val_df.shape[0]
```

## 2.2. Merge train set

```python
# Merge news_2023_df and news_2024_df before 2024-12-09 along column_names
merged_train_df = pd.concat([news_2023_df[merge_columns], news_2024_df[news_2024_df['date'] < pd.to_datetime("2024-12-07").date()][merge_columns]], axis=0)
merged_train_df.dropna(subset=['sentiment'], ignore_index=True, inplace=True)
merged_train_df.drop_duplicates(subset=['annot_title'], inplace=True) # Drop again just to be sure
merged_train_df = merged_train_df.sort_values(by="date", ascending=True) # Sort by date in ascending order
merged_train_df.reset_index(drop=True, inplace=True)
merged_train_df.head()
```

```python
# Save the last ~2.5 months as validation set for sentiment classification and the rest as train set
merged_train_df.to_csv('data/merged/news_SA_train.csv', index=False, encoding="utf-8")
merged_train_df.shape[0]
```
